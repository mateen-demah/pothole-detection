```python
import csv
from os import listdir
import numpy as np
from matplotlib import pyplot as plt
from importlib import reload
plt = reload(plt)
```

```python
with open('./raw/real1_sensordata.csv', 'r', encoding='utf-8', newline='') as sensordata_file:
    sensordata_file.readline() # skip headers
    sensordata = csv.reader(sensordata_file)
    time_accy = [[float(row[0]), float(row[4])] for row in sensordata]
```

```python
time_accy = np.array(time_accy)
time_accy[:, 0] -= time_accy[0,0]
print(time_accy[0,0])
print(time_accy.shape)
print(time_accy[0:2, :])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(time_accy[:20,0], time_accy[:20, 1], linewidth=5.5)
plt.xlabel("timestamp")
plt.ylabel("accelerometer y")
plt.show()
```
